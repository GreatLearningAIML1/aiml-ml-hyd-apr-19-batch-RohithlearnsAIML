```python
#function accepts two integers Goalsforinteger and goalsagainstinteger and returns gfi-gai
def Goal_diff_count(GFI,GAI):
    return GFI-GAI

```

```python
#create a new dataframe and copy loliga
GoalDiff=Laliga.copy()
```

```python
#Create two new columns GFI , GAI and copy the values from GoalsFOr and GoalsAgainsr and convert into int
GoalDiff["GFI"]=GoalDiff["GoalsFor"].astype('int')
GoalDiff["GAI"]=GoalDiff["GoalsAgainst"].astype('int')
```

```python
#Create a new column difference and pass the values of GFI and GAI into goal_diff_count function to calculate the difference of goals
GoalDiff["difference"]=Goal_diff_count(GoalDiff["GFI"],GoalDiff["GAI"])
```

```python
#use idxmax to find out the index and get the location and use iloc to get the team name
MaxGoalDiff=GoalDiff.iloc[GoalDiff["difference"].idxmax()]
MaxGoalDiff["Team"]
```

```python
#use idxmin to find out the index and get the location and use iloc to get the team name
MinGoalDiff=GoalDiff.iloc[GoalDiff["difference"].idxmin()]
MinGoalDiff["Team"]
```

```python
winning["GP"]=winning["GamesPlayed"].astype('int')
```

```python
def Win_Perc(GW,GP):
    return (GW/GP)*100



```

```python
winning["Winning Percent"]=Win_Perc(winning["GW"],winning["GP"])
```

```python
#finding the winning percent and filling the numerical errors with 0%
winning["Winning Percent"].fillna(0)
```

```python
#Creating new column winning Percent and appending it to the dataset Laliga
Laliga["Winning Percent"]=winning["Winning Percent"].astype('float')
```
